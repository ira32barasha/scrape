```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException



from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from random import  shuffle
import pandas as pd
import time
import dbm
import sqlite3
import json
import csv
```

```python
firefoxOptions = Options()

driver= webdriver.Firefox(executable_path="./selenium-firefox/drivers/geckodriver", options=firefoxOptions)

firefoxOptions.add_argument("-headless")


urls=list()


#urls=['https://www.amazon.de/dp/000420736X','https://www.amazon.fr/dp/000420736X','https://www.amazon.de/dp/000004458X','https://www.amazon.fr/dp/000004458X','https://www.amazon.de/dp/000101742X','https://www.amazon.fr/dp/000101742X']
with open('AmazonSheet1.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #https://www.amazon.{country}/dp/{asin}
        g = "https://www.amazon."+row["country"]+"/dp/"+row["Asin"]
        urls.append(g)

# Shuffling to avoid being detected by Amazon
shuffle(urls)



# Getting the start time to track on time required
start = time.time()
product_list= list()
k=0

# -------------------------------Web Scraping-------------------------------
for link in urls:
    # Open the url
    driver.get(link)
    job_dict = {}
    
    k=k+1
    print("####################",k,"##################")
    timeout = 3
    try:
        WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.ID, "productTitle")))
        
    except TimeoutException:
        print('Timed out: URL not available ')
        continue

      
    # -------------------------------Product title-------------------------------
    # find_elements_by_id returns an array of selenium objects.
    product_title = driver.find_element(By.ID,'productTitle').text;
    print("product title:",product_title)
    
    job_dict['product_title'] = product_title    
    print(link)  
    # -------------------------------Product Image URL-------------------------------
    try:
        Product_Image_URL = driver.find_element(By.XPATH,'//*[@id="landingImage"]').get_attribute('src')
        
    except:
        Product_Image_URL = driver.find_element(By.XPATH,'//*[@id="imgBlkFront"]').get_attribute('src')
    print("product Image URL",Product_Image_URL)
    job_dict['Product_Image_URL'] = Product_Image_URL
    
   
    # -------------------------------Product price-------------------------------
    # This will return the product price, if product price of priceblock is not found, move on to the other element
    
    
    listXPath = ['.//span[@class ="a-color-base"]' , './/ul/li/span/span[1]/span/a/span[2]/span' , './/div[11]/div[1]/div[1]/span[1]/span[2]']  
    
    for i in range(len(listXPath)):
        xPath = listXPath[i]
        try:
            #find_issuer = driver.find_element_by_xpath(xpath_issuer)
            product_price = driver.find_element_by_xpath(xPath).text
            
            
            
        except:
            pass 
    job_dict['product_price'] = product_price
    print("product price ",product_price)
    
    
    
    # -------------------------------Product Details     -------------------------------
    pdXPath = ['//*[@id="detailBullets_feature_div"]/ul','//div[34]/div/ul/li/span','//*[@id="detailBulletsWrapper_feature_div"]']
    
    for i in range(len(pdXPath)):
        xpath = pdXPath[i]
        try:
            #find_issuer = driver.find_element_by_xpath(xpath_issuer)
            product_details = driver.find_element_by_xpath(xpath).text
            
            
            
        except:
            pass 
    job_dict['product_details'] = product_details
    print("product details ",product_details)
    
    
    product_list.append(job_dict) 
    
    

    while k > 99:
        # Generate time tracker print
        end = time.time()
        print('for end-start after 100 urls')
        print("The time taken is {}", end-start)
        start = time.time()
        k=0
        break

        
    


# --- after loop ---
f = open('jsonfile.json', 'w')
#f.write(json.dumps(all_jobs)) # all in one line
f.write(json.dumps(product_list, indent=4))
f.close()

```

```python
################## saving data inside sqlite3 database ###################
import sqlite3
conn = sqlite3.connect('data.db', detect_types=sqlite3.PARSE_DECLTYPES)

cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS table1
           (product_title VARCHAR,
           Product_Image_URL   VARCHAR,
           product_price    VARCHAR,
           product_details VARCHAR)''')
try:
  
    query = "insert into table1 (product_title, Product_Image_URL, product_price, product_details) VALUES (?, ?, ?, ?)"
    for i in range(0, len(product_list)):
        cursor.executemany(query, ((product_list[i]["product_title"], product_list[i]["Product_Image_URL"], product_list[i]["product_price"], product_list[i]["product_details"]),))
    cursor.connection.commit()
    print("Multiple inserton operation done")
    cursor.close()
except sqlite3.Error as e:
    print("Error while inserting multiple data",e)
finally:
    if(conn):
        conn.close()
        print("Connection closed")
  
```
